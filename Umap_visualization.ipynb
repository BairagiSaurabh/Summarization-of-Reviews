```python
import pandas as pd
import pickle
import numpy as np
import umap.plot  # pip install umap-learn[plot]
from sklearn.preprocessing import OrdinalEncoder
```

```python
data = pd.read_csv("/content/drive/MyDrive/Project/neuralcoref/Viz_data.csv")
```

```python
data
```

```python
sentence_vectors = pickle.load(open('/content/drive/MyDrive/Project/neuralcoref/review_embedding','rb'))
```

```python
sentence_vectors.shape
```

```python
ord_enc = OrdinalEncoder()
y_encoded = ord_enc.fit_transform(data["Aspect"].values.reshape(-1,1))
```

```python
manifold = umap.UMAP().fit(sentence_vectors,y_encoded)
```

```python
umap.plot.points(manifold, labels=data["Aspect"].values, theme="fire")
```
